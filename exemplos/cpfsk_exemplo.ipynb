```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.signal import upfirdn
```

```python
L = 50 # oversampling factor 
Tb = 0.5 # bit period in seconds 
fs = L/Tb # sampling frequency in Hertz 
fc = 2/Tb # carrier frequency 
N = 8 # number of bits to transmit 
h = 1 # modulation index
```

```python
b = 2*np.random.randint(2, size=N)-1 # random information sequence
b_up = upfirdn(h = np.ones(L),x =b ,up = L)
b_integrated = np.convolve(b_up,np.ones(L))/fs # Integral approximation
theta= np.pi*h/Tb*b_integrated 
t=np.arange(0, Tb*N, 1/fs) # time base
```

```python
plt.plot(t,b_up)
plt.show()
```

```python
t2=np.arange(0,  len(b_integrated)/fs, 1/fs)
plt.plot(t2,theta)
plt.show()
```

```python
s = np.cos(2*np.pi*fc*t2 + theta) # CPFSK signal

plt.plot(t2,s)
plt.show()
```
